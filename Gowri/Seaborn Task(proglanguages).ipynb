```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df=pd.read_csv('proglanguages.csv')
df.head()
```

```python
Id=(np.asarray(df['id'])).reshape(5,8)
val=(np.asarray(df['value'])).reshape(5,8)
par=(np.asarray(df['parent'])).reshape(5,8)
rank=(np.asarray(df['rank'])).reshape(5,8)

print(Id)
print(val)
print(par)
print(rank)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='value')
result
```

```python
labels=(np.array(["{0} \n {1} \n {2} \n {3}".format(id,val,P,R)
                 for id,val,P,R in zip(Id.flatten(),
                                      val.flatten(),
                                      par.flatten(),
                                      rank.flatten())])).reshape(4,10)
labels
```

```python
fig,ax=plt.subplots(figsize=(22,12))

title="Proglanguages"

plt.title(title,fontsize=20)
ttl=ax.title
ttl.set_position([0.5,1.05])


ax.set_xticks([])
ax.set_yticks([])

ax.axis('off')

im=ax.imshow(result)
fig.colorbar(im)
               

#Use the heatmap function from the seaborn package
sn.heatmap(result,annot=labels,fmt="",cmap='icefire_r',linewidths=0.30,linecolor='blue',ax=ax) #result is pivot value

#Display 
plt.show()
```
