```python
import numpy as np
import pandas as pd
import seaborn as sn #seaborn id for data visualization
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('nifty50.csv')
df.head()
```

```python
df.columns
```

```python
df.drop(['Low', 'Last Traded Price', 'Change', 'Traded Volume(lacs)', 'Traded Value(crs)', '52 Week High',
       '52 Week Low', '365 Days % Change', '30 Days % Change'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"%Change":"Change"},inplace=True)
#df.replace({"%Change":"Change"},regex=True)

df.head()
```

```python
symbol=(np.asarray(df['Symbol'])).reshape(5,10)
perchange=(np.asarray(df['Change'])).reshape(5,10)
open=(np.asarray(df['Open'])).reshape(5,10)
high=(np.asarray(df['High'])).reshape(5,10)

print(symbol)
print(perchange)
print(open)
print(high)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='Change')
print(result)
```

```python
labels=(np.array(["{0} \n {1:.2f} \n {2} \n {3}".format(symb,value,O,H) 
                  for symb,value,O,H in zip(symbol.flatten(),
                    perchange.flatten(),
                    open.flatten(),
                high.flatten())])).reshape(10,5) #symb, values are run time variables calling from function
labels
```

```python
fig,ax=plt.subplots(figsize=(20,10))

#Add title to the heat map
title="Nifty 50" #title is a variable to print title we use 'plt'

#Set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

#Hide ticks for X & Y axis
ax.set_xticks([])
ax.set_yticks([])

#Remove the axes
ax.axis('off')

#Use the heatmap function from the seaborn package
sn.heatmap(result,annot=labels,fmt="",cmap='RdYlGn',linewidths=0.30,ax=ax) #result is pivot value

#Display the Nifty 50
plt.show()
```
