```python
##  creating a dataframe by extracting the csv file into df
df=pd.read_csv('cars93.csv')
```

```python
df
```

```python
# to see the all column names/heading
df.columns
```

```python
## dropping the unwanted columns from the dataframe
df.drop(['Manufacturer', 'Model', 'Type', 'Min.Price', 'Max.Price','MPG.city', 'MPG.highway', 'AirBags', 'DriveTrain', 'Cylinders','EngineSize', 'Rev.per.mile', 'Man.trans.avail', 'Passengers','Length', 'Wheelbase', 'Width','Turn.circle', 'Rear.seat.room', 'Luggage.room', 'Weight', 'Origin',],axis=1,inplace=True)
```

```python
df.head()
```

```python
## converting the columns into array
horsepower=(np.asarray(df['Horsepower'])).reshape(6,15)
make=(np.asarray(df['Make'])).reshape(6,15)
rpm=(np.asarray(df['RPM'])).reshape(6,15)
price=(np.asarray(df['Price'])).reshape(6,15)

print(horsepower)
print(make)
print(rpm)
print(price)
```

```python
# creating a pivot table using the dataframe
result=df.pivot(index='Yrows',columns='Xcols',values='Price')
result
```

```python
# to get the values and names in the box we do labels
labels=(np.asarray(["{0} \n {1} \n {2} \n {3}".format(symb,horse,speed,rate) for symb,horse,speed,rate in zip(make.flatten(),horsepower.flatten(),rpm.flatten(),price.flatten())])).reshape(15,6)
labels
```

```python
# Defining the plot
fig,ax=plt.subplots(figsize=(20,15))

# add title to the heat map
title ="Nifty 50"

# setthe font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

# hide ticks for x and y axis
ax.set_xticks([])
ax.set_yticks([])

# remove the axes
ax.axis('off')

#use the heatmap function from  the seaborn package
sn.heatmap(result,annot=labels,fmt="",cmap='RdYlGn',linewidths=0.30,ax=ax)

# display the nifty 50
plt.show()
```
