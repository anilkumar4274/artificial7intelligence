```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import os
os.chdir('C:\\Users\\Naveen Kumar\\artificial7intelligence\\projects\\datasets')
df=pd.read_csv('ozone.csv')
```

```python
df=df[0:200]
```

```python
df.shape
```

```python
def xy(r):
    d=[]
    global a,b
    for i in range(1,r+1):
        if(r%i==0):
            d.append(i)
    if len(d)%2==0:
        a=d[int(len(d)/2-1)]
        b=d[int(len(d)/2)]
    df1=pd.DataFrame(index=list(np.arange(r)),columns={'Yrows','Xcols'})
    j,k=0,0
    while(k<=b-1):
        for i in range(b):
            df1['Yrows'][j]=k
            df1['Xcols'][j]=i
            j=j+1
        k=k+1
    return(df1)
```

```python
df=df.join(xy(len(df)))
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['Month', 'Day_of_month', 'Day_of_week', 'Wind_speed','Temperature_ElMonte', 'Inversion_base_height','Inversion_temperature', 'Visibility'],axis=1,inplace=True)
```

```python
df.columns
```

```python
pressure=(np.asarray(df['pressure_height'])).reshape(a,b)
humidity=(np.asarray(df['Humidity'])).reshape(a,b)
P_gradient=(np.asarray(df['Pressure_gradient'])).reshape(a,b)
ozone=(np.asarray(df['ozone_reading'])).reshape(a,b)
result=df.pivot(index='Yrows',columns='Xcols',values='ozone_reading')
labels=(np.asarray(["{0} \n {1} \n {2} \n {3:.2f}".format(pre,humi,grade,ozo) for pre,humi,grade,ozo in zip(pressure.flatten(),humidity.flatten(),P_gradient.flatten(),ozone.flatten())])).reshape(a,b)
```

```python
fig,ax=plt.subplots(figsize=([25,15]))


#Set the font size and the distance of the title from the plot
plt.title('OZONE',fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])


ax.set_xticks([])
ax.set_yticks([])

#remove the axis
ax.axis('off')
#Use the heatmap function from the seaborn package

sn.heatmap(result,annot=labels,fmt="",cmap="RdYlGn",linewidth=0.30,square=True,ax=ax)

#display final
plt.show()
```
