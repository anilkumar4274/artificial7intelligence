```python
for i in range(len(df)):
    df["Scrip"][i] = df["Scrip"][i][:-3:]
df.head()
```

```python
sl = []
cl = []

for i in range(len(df)):
    l = len(df["Scrip"][i])
    for j in range(len(bse_df)):
        if df["Scrip"][i]==bse_df["STOCK"][j][:l:]:
            sl.append(bse_df["STOCK"][j][:l:])
            cl.append(bse_df["CODE"][j])
            break
```

```python
res = pd.DataFrame(columns = ["stock","code"])
res["stock"] = sl
res["code"] = cl
```
