```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('mtcars.csv')
df.columns
```

```python
df.head()
```

```python
df.drop(['mpg','drat', 'wt', 'qsec', 'vs', 'am', 'gear',
       'carb', 'fast', 'cars'],axis=1,inplace=True)
```

```python
carname=(np.asarray(df['carname'])).reshape(8,4)
cylinders=(np.asarray(df['cyl'])).reshape(8,4)
displacement=(np.asarray(df['disp'])).reshape(8,4)
horsepower=(np.asarray(df['hp'])).reshape(8,4)

print(carname)
print(cylinders)
print(displacement)
print(horsepower)
```

```python
# creating a pivot table using the dataframe
result=df.pivot(index='Yrows',columns='Xcols',values='hp')
result
```

```python
# to get the values and names in the box we do labels
labels=(np.asarray(["{0} \n {1} \n {2} \n {3}".format(carname,cylinders,displacement,horsepower) for carname,cylinders,displacement,horsepower in zip(carname.flatten(),cylinders.flatten(),displacement.flatten(),horsepower.flatten())])).reshape(8,4)
labels
```

```python
#Defining the plot
fig,ax=plt.subplots(figsize=(20,15))

# add title to the heat map
title ="cars"

# setthe font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

# hide ticks for x and y axis
ax.set_xticks([])
ax.set_yticks([])

# remove the axes
ax.axis('off')

#use the heatmap function from  the seaborn package
sns.heatmap(result,annot=labels,fmt="",cmap='RdYlGn',linewidths=0.30,ax=ax)

# display the nifty 50
plt.show()
```
