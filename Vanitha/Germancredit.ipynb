```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as snss
```

```python
df=pd.read_csv('GermanCredit.csv')
```

```python
df.head()
```

```python
print(df.shape)
df.columns
```

```python
df.drop(['savings','employment_duration', 'installment_rate', 'personal_status_sex','other_debtors', 'present_residence', 'property', 'age','other_installment_plans', 'housing', 'number_credits', 'job','people_liable', 'telephone', 'foreign_worker', 'credit_risk', 'credit_history'],axis=1,inplace=True)
```

```python
df.head()
```

```python
Status=(np.asarray(df['status'])).reshape(9,10);Status
```

```python
Duration=(np.asarray(df['duration'])).reshape(9,10);Duration
```

```python
Purpose=(np.asarray(df['purpose'])).reshape(9,10);Purpose
```

```python
Amount=(np.asarray(df['amount'])).reshape(9,10);Amount
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='amount');result
```

```python
labels=(np.asarray(["{0} \n {1} \n {2} \n {3}".format(sta,dura,pur,amo) 
                   for sta,dura,pur,amo in zip(Status.flatten(),Duration.flatten(),Purpose.flatten(),Amount.flatten())])).reshape(18,5,)
labels
```

```python
fig,ax=plt.subplots(figsize=(20,19))

#ADD TITLE TO THE HEAT MAP
title="GermanCredit"

#SET THE FONT SIZE AND THE DIsTANCE OF THE TITLE FROM THE plOT
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

#HIDE TICKS FOR X AND Y AXIS
ax.set_xticks([])
ax.set_yticks([])

#REMOVE THE AXIS
ax.axis('off')
#fig,ax=plt.subplots(figsize=(12,12))
snss.heatmap(result,annot=labels,fmt="",cmap='gist_rainbow',linewidths=0.30,linecolor='green',ax=ax)
plt.show()
```
