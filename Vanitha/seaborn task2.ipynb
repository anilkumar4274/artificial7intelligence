```python
import pandas as pd
import numpy as np
import seaborn as snss
from sklearn.preprocessing import Imputer
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Cars93_miss.csv')
```

```python
df
```

```python
df.shape
```

```python
import math 
med_Price=math.floor(df.Price.median())
med_Price
```

```python
df.Price.median()
```

```python
df.Price=df.Price.fillna(med_Price)
df
```

```python
price=np.asarray(df['Price']).reshape(9,10)
print(price)
```

```python
max_price=np.asarray(df['Max.Price']).reshape(9,10)
print(max_price)
```

```python
manufacturer=np.asarray(df['Manufacturer']).reshape(9,10)
print(manufacturer)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='Price')
print(result)
```

```python
labels=np.asarray(["{0}\n {1:.2f}\n".format(manufac,price) for manufac,price in zip(manufacturer.flatten(),price.flatten())]).reshape(9,10)
print(labels)
```

```python
fig,ax=plt.subplots(figsize=(19,8))

#ADD TITLE TO THE HEAT MAP
title="Cars93_miss"

#SET THE FONT SIZE AND THE DIsTANCE OF THE TITLE FROM THE plOT
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

#HIDE TICKS FOR X AND Y AXIS
ax.set_xticks([])
ax.set_yticks([])

#REMOVE THE AXIS
ax.axis('off')
#fig,ax=plt.subplots(figsize=(12,12))

#im=ax.imshow(result)
#fig.colorbar(im)
#for (j,i), x in np.ndenumerate(result):
 #   url = "https://forge.scilab.org/index.php/p/rdataset/source/tree/master/csv/MASS/Cars93.csv"
  #  ax.annotate(x, xy=(i,j), ha="center", va="center",
   #             url=url, bbox=dict(color='r', alpha=1e-6, url=url)) #ha and va= horizontal and vertical axis.

#USE THE HEAT  MAP FUNCTION FROM THE SEABORN PACKAGE
snss.heatmap(result,annot=labels,fmt="",cmap='spring',linewidths=0.30,linecolor='green',ax=ax)
plt.show()

#annotate = checkk syntax to know the explanation.                 
```
