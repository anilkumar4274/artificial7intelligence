```python
import pandas as pd
import numpy as np
import seaborn as snss
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('diamonds.csv')
```

```python
df.drop(['carat','color','depth','table','y'],axis=1,inplace=True)
```

```python
df.head()
```

```python
Cut=(np.asarray(df['cut'])).reshape(9,10)
```

```python
Cut
```

```python
Price=(np.asarray(df['price'])).reshape(9,10)
```

```python
Price
```

```python
X=(np.asarray(df['x'])).reshape(9,10)
```

```python
X
```

```python
Z=(np.asarray(df['z'])).reshape(9,10)
```

```python
Z
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='price')
print(result)
```

```python
labels=(np.asarray(["{0} \n {1} \n {2} \n {3}".format(pri,cu,x,z) 
                   for pri,cu,x,z in zip(Price.flatten(),Cut.flatten(),X.flatten(),Z.flatten())])).reshape(18,5)
labels
```

```python
fig,ax=plt.subplots(figsize=(20,18))

#ADD TITLE TO THE HEAT MAP
title="diamonds"

#SET THE FONT SIZE AND THE DIsTANCE OF THE TITLE FROM THE plOT
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

#HIDE TICKS FOR X AND Y AXIS
ax.set_xticks([])
ax.set_yticks([])

#REMOVE THE AXIS
ax.axis('off')
#fig,ax=plt.subplots(figsize=(12,12))
snss.heatmap(result,annot=labels,fmt="",cmap='Wistia',linewidths=0.30,linecolor='green',ax=ax)
plt.show()

#annotate = checkk syntax to know the explanation.                 
```
