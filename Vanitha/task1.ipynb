```python
import numpy as np
import pandas as pd
import seaborn as snss
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('nifty50.csv')
df
```

```python
df.head()
```

```python
print(df.shape)
df.columns
```

```python
df.drop(['Last Traded Price', 'Change','Low','52 Week Low', '365 Days % Change', '30 Days % Change','Traded Volume(lacs)', 'Traded Value(crs)', '52 Week High'],axis=1,inplace=True)

df.head()
```

```python
df.rename(columns={"%Change":"Change"},inplace=True)
df.head()
```

```python
(np.asarray(df['Symbol'])).reshape(5,10)
```

```python
symbol=(np.asarray(df['Symbol'])).reshape(5,10)

print(symbol)
```

```python
perchange=(np.asarray(df['Change'])).reshape(5,10)
print(perchange)
```

```python
open=(np.asarray(df['Open'])).reshape(5,10)
print(open)
```

```python
high=(np.asarray(df['High'])).reshape(5,10)
print(high)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='Change')
print(result)
```

```python
labels=np.asarray(["{0}\n {1:.2f}\n {2}\n {3}".format(symb,value,ope,hig) for symb,value,ope,hig in zip(symbol.flatten(),perchange.flatten(),open.flatten(),high.flatten())]).reshape(10,5)
print(labels)
```

```python
#DEFINE THE PLOT
fig,ax=plt.subplots(figsize=(19,10))

#ADD TITLE TO THE HEAT MAP
title="Nifty 50"

#SET THE FONT SIZE AND THE DIsTANCE OF THE TITLE FROM THE plOT
plt.title(title,fontsize=20)
ttl=ax.title
ttl.set_position([0.5,1.05])

#HIDE TICKS FOR X AND Y AXIS
ax.set_xticks([])
ax.set_yticks([])

#REMOVE THE AXIS
ax.axis('off')
#USE THE HEAT  MAP FUNCTION FROM THE SEABORN PACKAGE
snss.heatmap(result,annot=labels,fmt="",cmap='RdYlGn',linewidths=0.50,linecolor='blue',ax=ax)
plt.show()
```
