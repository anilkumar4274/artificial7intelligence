```python
import pandas as pd
import glob, os
df = pd.concat(map(pd.read_csv, glob.glob(os.path.join('C://Users/mural/Desktop/imp/Python/New folder/banknifty/30/', "*.csv"))))
df=df.drop(["Expiry","Symbol","Settle Price","No. of contracts","Turnover in Lacs","Premium Turnover in Lacs","Open Int","Change in OI","Underlying Value"],axis=1)
def func(start_date,option_type='CE'):
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import os
    from datetime import datetime
    import os
    os.chdir('C://Users/mural/Desktop/imp/Python/New folder/banknifty/')
    global df_d
    df_d=pd.read_csv('data.csv')
    df_d=df_d.set_index(['Date'])
    df_d.index = pd.to_datetime(df_d.index)
    df_d.drop(['Shares Traded','Turnover (Rs. Cr)'],axis=1,inplace=True)
    df_d['CO']=df_d.Close-df_d.Open
    round_price=df_d.loc[start_date]['Close']
    x=np.round(round_price,decimals=-2)
    cp=[((x+100),(x-100)) for i in range(1)] 
    #import glob, os    
    #df = pd.concat(map(pd.read_csv, glob.glob(os.path.join('C://Users/mural/Desktop/imp/Python/New folder/banknifty/30/', "*.csv"))))
    #df=df.drop(["Expiry","Symbol","Settle Price","No. of contracts","Turnover in Lacs","Premium Turnover in Lacs","Open Int","Change in OI","Underlying Value"],axis=1)
    ddatetime_series = pd.to_datetime(df['Date'])
    ddatetime_index = pd.DatetimeIndex(ddatetime_series.values)
    df['Date']=ddatetime_index
    df['O-C']=df['Open']-df['Close']
    #df=df[df["Option Type"] == option_type ]
    #df.drop(['Expiry'],axis=1,inplace=True)
    if option_type == 'CE':
        dfc=df[df["Option Type"] == option_type ]
        a=dfc[(dfc['Date']==start_date) & (dfc['Strike Price']==cp[0][0])]
        print(a)
    elif option_type == 'PE':
        dfp=df[df["Option Type"] == option_type ]
        b=dfp[(dfp['Date']==start_date) & (dfp['Strike Price']==cp[0][1])]
        print(b)
    elif option_type == 'both':
        dfc=df[df["Option Type"] == 'CE' ]
        a=dfc[(dfc['Date']==start_date) & (dfc['Strike Price']==cp[0][0])]
        dfp=df[df["Option Type"] == 'PE' ]
        b=dfp[(dfp['Date']==start_date) & (dfp['Strike Price']==cp[0][1])]
        global y
        y=a['O-C'].sum()+b['O-C'].sum()
        return(y)
```

```python
df_bd=pd.DataFrame()
df_bd['B']=pd.bdate_range('2019-01-01', '2019-06-21',freq='B')
```
