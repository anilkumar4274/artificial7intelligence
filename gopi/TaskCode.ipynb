```python
import pandas as pd
```

```python
df=pd.read_csv('table.csv')
data = pd.read_csv('bsestocks.txt', sep="|", header=None)
data.columns=data.iloc[0]
data.drop(df.index[0],inplace=True)
data.head()
```

```python
df.head()
```

```python
df4=pd.DataFrame(columns=['CODE'])

for i in range(len(df)):
    k=len(df['0'][i])
    while(k>0):
        for j in range(1,len(data)):
            l=len(data['STOCK'][j])
            if(l>k):
                l=k
            if df['0'][i][:k].lower() == data['STOCK'][j].lower().replace(" ","")[:l]:
                df4=df4.append({'CODE':data['CODE'][j]},ignore_index=True)
                k=0
                break
        k=k-1
```

```python
df4.head()
```

```python
df['CODE']=df4
```
