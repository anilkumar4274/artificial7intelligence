```python
left=pd.DataFrame({'key1':['K0','K1','K2','K3'],
                   'key2':['K0','K1','K0','K1'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']
                     })
right=pd.DataFrame({'key1':['K0','K1','K2','K3'],
                    'key2':['K0','K0','K0','K0'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']
                   })
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=['key1','key2'])  ## only matching of common key1,key2 will appear at the merge columns(acc to columns)
```

```python
pd.merge(left,right,how='left',on=['key1','key2']) ## only matching part we will get ,otherwise nans will get.left with intersection
```

```python
pd.merge(left,right,how='right',on=['key1','key2']) ## where ever it is matching it will get values ,otherwise it will get nan values.right with intersection
```

```python
pd.merge(left,right,how='outer',on=['key1','key2']) ## union
```

```python
pd.merge(left,right,how='inner',on=['key1','key2']) ## common part ,intersection.
```
