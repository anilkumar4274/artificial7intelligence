```python
df = pd.read_csv('BostonHousing.csv')
df.columns
```

```python
df.drop(['chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat', ],axis=1,inplace=True)
```

```python
df.head()
```

```python
(np.asarray(df['indus'])).reshape(46,11)
```

```python
indus = (np.asarray(df['indus'])).reshape(46,11)
medv = (np.asarray(df['medv'])).reshape(46,11)

print(indus)
print(medv)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='indus')
print(result)
```

```python
labels = (np.asarray(["{0}\n{1:.2f}".format(indus,value)
                     for indus,value in zip(indus.flatten(),
                     medv.flatten())])).reshape(46,11)
```

```python
labels
```

```python
fig,ax=plt.subplots(figsize=(10,20))
title="BostonHousing"
plt.title(title,fontsize=25)
tt1=ax.title
tt1.set_position([0.5,1.05])
ax.set_xticks([])
ax.set_yticks([])
sns.heatmap(result,annot=labels,fmt="",cmap='spring_r',linewidths=0.50,ax=ax)
plt.show()
```
