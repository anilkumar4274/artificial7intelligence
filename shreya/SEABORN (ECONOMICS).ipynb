```python
df = pd.read_csv('economics.csv')
df.columns
```

```python
df.drop(['unemploy',],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['pce']
```

```python
(np.asarray(df['pce'])).reshape(55,10)
```

```python
pce = (np.asarray(df['pce'])).reshape(55,10)
psavert = (np.asarray(df['psavert'])).reshape(55,10)

print(pce)
print(psavert)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='pce')
print(result)
```

```python
labels = (np.asarray(["{0}\n{1:.2f}".format(pce,value)
                     for pce,value in zip(pce.ravel(),
                     psavert.ravel())])).reshape(55,10)
```

```python
labels
```

```python
fig,ax=plt.subplots(figsize=(30,40))
title="ECONOMICS"
plt.title(title,fontsize=25)
tt1=ax.title
tt1.set_position([0.5,1.05])
ax.set_xticks([])
ax.set_yticks([])
sns.heatmap(result,annot=labels,fmt="",cmap='spring_r',linewidths=0.50,ax=ax)
plt.show()
```
