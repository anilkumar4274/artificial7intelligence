```python
df = pd.read_csv('Hitters.csv')
df.columns
```

```python
df.drop(['RBI', 'Walks', 'Years', 'CAtBat',
       'CHits', 'CHmRun', 'CRuns', 'CRBI', 'CWalks', 'League', 'Division',
       'PutOuts', 'Assists', 'Errors', 'Salary',],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['AtBat']
```

```python
(np.asarray(df['AtBat'])).reshape(29,11)
```

```python
hits = (np.asarray(df['Hits'])).reshape(29,11)
runs = (np.asarray(df['Runs'])).reshape(29,11)

print(hits)
print(runs)
```

```python
result = df.pivot(index = 'Yrows',columns = 'Xcols',values = 'Runs')
print(result)
```

```python
labels = (np.asarray(["{0}\n{1:.2f}".format(hits,value)
                     for hits,value in zip(hits.flatten(),
                     runs.flatten())])).reshape(29,11)
```

```python
labels
```

```python
fig,ax=plt.subplots(figsize=(10,20))
title="Hitters"
plt.title(title,fontsize=25)
tt1=ax.title
tt1.set_position([0.5,1.05])
ax.set_xticks([])
ax.set_yticks([])
sns.heatmap(result,annot=labels,fmt="",cmap='spring_r',linewidths=0.50,ax=ax)
plt.show()
```
