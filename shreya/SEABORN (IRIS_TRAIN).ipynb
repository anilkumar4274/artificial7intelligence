```python
df = pd.read_csv('iris_train.csv')
df.columns
```

```python
df.rename(columns = {"Sepal.Length":"Length","Sepal.Width":"Width","Petal.Length":"Petal","Petal.Width":"Petal 1"},inplace=True)
```

```python
df.head()
```

```python
df['Length']
```

```python
(np.asarray(df['Width'])).reshape(35,3)
```

```python
petal = (np.asarray(df['Petal'])).reshape(35,3)
petal1 = (np.asarray(df['Petal 1'])).reshape(35,3)

print(petal)
print(petal1)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='Width')
print(result)
```

```python
labels = (np.asarray(["{0}\n{1:.2f}".format(petal,value)
                     for petal,value in zip(petal.flatten(),
                     petal1.flatten())])).reshape(35,3)
```

```python
labels
```

```python
fig,ax=plt.subplots(figsize=(10,20))
title="IRIS_TRAIN"
plt.title(title,fontsize=25)
tt1=ax.title
tt1.set_position([0.5,1.05])
ax.set_xticks([])
ax.set_yticks([])
sns.heatmap(result,annot=labels,fmt="",cmap='spring_r',linewidths=0.50,ax=ax)
plt.show()
```
