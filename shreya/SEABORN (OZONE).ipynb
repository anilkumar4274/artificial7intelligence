```python
df = pd.read_csv('ozone.csv')
df.columns
```

```python
df.drop(['Day_of_month', 'Day_of_week', 'ozone_reading',
       'pressure_height', 'Wind_speed', 'Humidity', 'Temperature_Sandburg',
       'Temperature_ElMonte',],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns = {"Month":"Mon","Inversion_base_height":"Height","Pressure_gradient":"Pressure","Inversion_temperature":"Temperature"},inplace=True)
```

```python
df.head()
```

```python
df['Height']
```

```python
(np.asarray(df['Height'])).reshape(33,6)
```

```python
height = (np.asarray(df['Height'])).reshape(33,6)
month = (np.asarray(df['Mon'])).reshape(33,6)

print(height)
print(month)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='Height')
print(result)
```

```python
labels = (np.asarray(["{0}\n{1:.2f}".format(height,value)
                     for height,value in zip(height.flatten(),
                     month.flatten())])).reshape(33,6)
```

```python
labels
```

```python
fig,ax=plt.subplots(figsize=(10,20))
title="Ozone"
plt.title(title,fontsize=25)
tt1=ax.title
tt1.set_position([0.5,1.05])
ax.set_xticks([])
ax.set_yticks([])
sns.heatmap(result,annot=labels,fmt="",cmap='spring_r',linewidths=0.50,ax=ax)
plt.show()
```
