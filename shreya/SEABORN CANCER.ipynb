```python
df = pd.read_csv('BreastCancer.csv')
df.columns
```

```python
df.drop(['Cell.shape', 'Marg.adhesion',
       'Epith.c.size', 'Bare.nuclei', 'Bl.cromatin',],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns = {"Cl.thickness":"Thickness","Cell.size":"Size","Normal.nucleoli":"Nucleoli"},inplace=True)
```

```python
df.head()
```

```python
df['Thickness']
```

```python
(np.asarray(df['Thickness'])).reshape(30,20)
```

```python
size = (np.asarray(df['Size'])).reshape(30,20)
nucleoli = (np.asarray(df['Nucleoli'])).reshape(30,20)

print(size)
print(nucleoli)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='Mitoses')
print(result)
```

```python
labels = (np.asarray(["{0}\n{1:.2f}".format(nucleoli,value)
                     for nucleoli,value in zip(nucleoli.flatten(),
                     size.flatten())])).reshape(30,20)
```

```python
labels
```

```python
fig,ax=plt.subplots(figsize=(10,20))
title="BreastCancer"
plt.title(title,fontsize=25)
tt1=ax.title
tt1.set_position([0.5,1.05])
ax.set_xticks([])
ax.set_yticks([])
sns.heatmap(result,annot=labels,fmt="",cmap='spring_r',linewidths=0.50,ax=ax)
plt.show()
```
