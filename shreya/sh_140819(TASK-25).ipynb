```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('nifty50.csv')
df.head()
```

```python
df.columns
```

```python
df.drop(['High', 'Low', 'Change','Traded Volume(lacs)', 'Traded Value(crs)', '52 Week High', '52 Week Low', '365 Days % Change', '30 Days % Change'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"%Change":"Change","Last Traded Price":"LTP"},inplace=True)
```

```python
symbol=(np.asarray(df['Symbol'])).reshape(5,10)
perchange=(np.asarray(df['Change'])).reshape(5,10)
opn=(np.asarray(df['Open'])).reshape(5,10)
ltp=(np.asarray(df['LTP'])).reshape(5,10)



print(ltp)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='Change')
print(result)
```

```python
##create an array to annotate the heatmap
labels=(np.asarray(["{0} \n {1:2f} \n {2} \n {3}".format(symb,value,opn,ltp)
                    for symb,value,opn,ltp in zip(symbol.flatten(),
                    perchange.flatten(),opn.flatten(),ltp.flatten())])).reshape(10,5)
labels
```

```python
##define the plot
fig,ax=plt.subplots(figsize=(10,20))
##add title to the heat map
title="Nifty 50"
##set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

##hide ticks for x&y axis.
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
#use the heatmap function from the seaborn package
sns.heatmap(result,annot=labels,fmt="",cmap='RdYlGn',linewidth=2,ax=ax)
##display the nifty 50
plt.show()

#fig, ax = plt.subplots()
im=ax.imshow(result)

fig.colorbar(im)
for(j,i), x in np.ndenumerate(result):
    url = "https://www.wolframalpha.com/input/?i={}".format(x)
    ax.annotate(x,xy=(i,j),ha="center",va="center",
                url=url,bbox=dict(color='w',alpha=1e-6,url=url))
```
