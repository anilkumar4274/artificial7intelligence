```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Auto.csv')
df.columns
```

```python
df.drop(['displacement', 'horsepower', 
       'acceleration'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.rename(columns={"cylinders":"cylinder"},inplace=True)
```

```python
df
```

```python
(np.asarray(df['name'])).reshape(5,10)
```

```python
mpg=(np.asarray(df['mpg'])).reshape(5,10)
cylinder=(np.asarray(df['cylinder'])).reshape(5,10)
year=(np.asarray(df['year'])).reshape(5,10)
name=(np.asarray(df['name'])).reshape(5,10)

print(mpg)
print(cylinder)
print(year)
print(name)
```

```python
# creating a pivot table using the dataframe
result=df.pivot(index='Yrows',columns='Xcols',values='mpg')
result
```

```python
# to get the values and names in the box we do labels
labels=(np.asarray(["{0} \n {1} \n {2} \n {3}".format(mpg,cylinder,year,name) for mpg,cylinder,year,name in zip(mpg.flatten(),cylinder.flatten(),year.flatten(),name.flatten())])).reshape(5,10)
labels
```

```python
#define the plot
fig,ax=plt.subplots(figsize=(9,15))


#Add title to the heat map
title="Auto"

#set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

#Hide the ticks for the x & y axis
ax.set_xticks([])
ax.set_yticks([])

#remove the axes
ax.axis('Auto')
#use the heat map function from sea born package
sns.heatmap(result,annot=True,fmt="",cmap='Oranges',linewidths=0.30,ax=ax)

#Display the species
plt.show()


```
