```python
import numpy as np
import pandas as pd
import seaborn as snss
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('nifty50.csv')
df.head()
```

```python
df.drop(['Low','Last Traded Price','Change','Traded Volume(lacs)','Traded Value(crs)','52 Week High',
         '52 Week Low','365 Days % Change','30 Days % Change'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename({'%Change':'change'},axis=1,inplace=True)
```

```python
df.head()
```

```python
#(np.asarray(df['Symbol']))  To convert series into array
symbol=(np.asarray(df['Symbol'])).reshape(5,10)
change=(np.asarray(df['change'])).reshape(5,10)
Open=(np.asarray(df['Open'])).reshape(5,10)
High=(np.asarray(df['High'])).reshape(5,10)
print(symbol)
print(change)
print(Open)
print(High)
```

```python
result1=df.pivot(index='Yrows',columns='Xcols',values='change')
print(result1)
```

```python
#Create an array to annotate the heatmap
labels=(np.asarray(['{0} \n {1:.2f} \n {2} \n {3}'.format(symb,value,Open,High)    #format:it helps outside characters can print as inside characters
                    for symb,value,Open,High in zip(symbol.flatten(),
                                          change.flatten(),Open.flatten(),High.flatten())])).reshape(10,5)   #flatten:it can change 2D to single dimension
labels
```

```python
#Define the plot
fig,ax=plt.subplots(figsize=(17,9))  #to plot a  outer grid which size we want i.e.,(length,width)

#Add title to the Heat map
title="Nifty 50"   

#Set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)    #to display the title and title fontsize is ourwish
tt1=ax.title
tt1.set_position([0.5,1.05])   #title position setting

#Hide ticks for x & y axis
ax.set_xticks([])
ax.set_yticks([])

#Remove the axes
ax.axis('off')  #total axes creation will be sent to background

#Use the Heatmap function from the Seaborn package
snss.heatmap(result1,annot=labels,fmt="",cmap="RdYlGn",linewidths=0.30,ax=ax)   #cmap:color map,result=our created data,fmt(format)
#Display the Nifty 50                          Reddark,yellowlight,Green normal
plt.show()
```
