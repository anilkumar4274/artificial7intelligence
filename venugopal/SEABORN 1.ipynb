```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month","year","passengers")
ax = sns.heatmap(flights)
```

```python
ax = sns.heatmap(flights, annot=False, fmt="d")
```

```python
ax = sns.heatmap(flights, linewidth=.5, fmt="d",annot=True)
```

```python
ax = sns.heatmap(flights, linewidth=.5, fmt="d",annot=False)
```

```python
grid_kws = {"height_ratios": (.9, .05), "hspace": .3}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(flights, ax=ax,
cbar_ax=cbar_ax,
cbar_kws={"orientation": "horizontal"})
```

```python
grid_kws = {"height_ratios": (.5, .05), "hspace": .3}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(flights, ax=ax,
cbar_ax=cbar_ax,
cbar_kws={"orientation": "vertical"})
```

```python
corr = np.corrcoef(np.random.randn(10,200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```

```python
corr = np.corrcoef(np.random.randn(20, 500))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("dark"):
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```

```python
corr = np.corrcoef(np.random.randn(10, 200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("whitegrid"):
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```

```python
corr = np.corrcoef(np.random.randn(10, 200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("darkgrid"):
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```

```python
corr = np.corrcoef(np.random.randn(10, 200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("ticks"):
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```

```python
corr = np.corrcoef(np.random.randn(10, 200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = False
with sns.axes_style("ticks"):
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=False)
```
