```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('nifty50.csv')
df.columns
```

```python
df.drop(['Last Traded Price', 'Change', 'Low','Traded Volume(lacs)', 'Traded Value(crs)', '52 Week High',
       '52 Week Low', '365 Days % Change', '30 Days % Change'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"%Change":"Change"},inplace=True)
```

```python
df.head()
```

```python
(np.asarray(df['Symbol'])).reshape(5,10)
```

```python
symbol=(np.asarray(df['Symbol'])).reshape(5,10)
perchange=(np.asarray(df['Change'])).reshape(5,10)
Open=(np.asarray(df['Open'])).reshape(5,10)
High=(np.asarray(df['High'])).reshape(5,10)

print(symbol)
print(perchange)
print(Open)
print(High)
```

```python
result=df.pivot(index='Yrows',columns='Xcols',values='Change')
print(result)
```

```python
labels=(np.asarray(["{0} \n {1:2f} \n {2} \n {3}".format(symb,value,Open,High)
                   for symb,value,Open,High in zip(symbol.flatten(),
                                          perchange.flatten(),Open.flatten(),High.flatten())])).reshape(10,5)   
labels
```

```python
#define the plot
fig,ax=plt.subplots(figsize=(20,10))


#Add title to the heat map
title="Nifty 50"

#set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

#Hide the ticks for the x & y axis
ax.set_xticks([])
ax.set_yticks([])

#remove the axes
ax.axis('off')
#use the heat map function from sea born package
sns.heatmap(result,annot=labels,fmt="",cmap='RdYlGn',linewidths=0.30,ax=ax)

#Display the Nifty 50
plt.show()


```
