```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris_train1.csv')
df.columns
```

```python
df.head()
```

```python
df.rename(columns={"Sepal.Length":"sepallength","Sepal.Width":"sepalwidth","Petal.Length":"petallength","Petal.Width":"petalwidth"},inplace=True)
```

```python
df.head()
```

```python
sepallength=(np.asarray(df['sepallength'])).reshape(11,5)
```

```python
sepallength=(np.asarray(df['sepallength'])).reshape(11,5)
sepalwidth=(np.asarray(df['sepalwidth'])).reshape(11,5)
petallength=(np.asarray(df['petallength'])).reshape(11,5)
petalwidth=(np.asarray(df['petalwidth'])).reshape(11,5)

print(sepallength)
print(sepalwidth)
print(petallength)
print(petalwidth)
```

```python
# creating a pivot table using the dataframe
result=df.pivot(index='Yrows',columns='Xcolumns',values='petalwidth')
result
```

```python
# to get the values and names in the box we do labels
labels=(np.asarray(["{0} \n {1} \n {2} \n {3}".format(sepallength,sepalwidth,petallength,petalwidth) for sepallength,sepalwidth,petallength,petalwidth in zip(sepallength.flatten(),sepalwidth.flatten(),petallength.flatten(),petalwidth.flatten())])).reshape(11,5)
labels
```

```python
#define the plot
fig,ax=plt.subplots(figsize=(9,15))


#Add title to the heat map
title="species"

#set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl=ax.title
ttl.set_position([0.5,1.05])

#Hide the ticks for the x & y axis
ax.set_xticks([])
ax.set_yticks([])

#remove the axes
ax.axis('off')
#use the heat map function from sea born package
sns.heatmap(result,annot=labels,fmt="",cmap='ocean_r',linewidths=0.30,ax=ax)

#Display the species
plt.show()


```
